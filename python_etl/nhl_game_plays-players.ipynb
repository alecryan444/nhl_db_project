```python
import pandas as pd
import requests 
import sqlalchemy

pd.set_option('display.max_columns', 500)
```

```python
##Game ID
r = requests.get('https://statsapi.web.nhl.com/api/v1/schedule?season=20192020')
j = r.json()

game_list = []

for date in j['dates']:
    dt = date['date']
    for game in date['games']:
        game_num = game['gamePk']
        if game['gameType'] != 'PR':
            game_list.append(game_num)
        else:
            pass
```

```python
df_list = []

#game_list = [' 2019020920']

for game in game_list:
    r = requests.get(f"https://statsapi.web.nhl.com/api/v1/game/{game}/feed/live")
    j = r.json()
    
    print(game)
    try:
        play_dict = {}
        for i in list(range(len(j['liveData']['plays']['allPlays']))):


                game_type = j['gameData']['game']['type'] 


                for key, val in j['liveData']['plays']['allPlays'][i].items():
                    event_id = j['liveData']['plays']['allPlays'][i]['about']['eventIdx']
                    event_code = j['liveData']['plays']['allPlays'][i]['result']['eventCode']
                    
                    if key == 'players':
                        for player in val:
                            pt = player['playerType']
                            p_id = player['player']['id']
                            fn = player['player']['fullName']
                            play_dict.setdefault(f'playerType', []).append(pt)
                            play_dict.setdefault(f'playerId', []).append(p_id)
                            play_dict.setdefault(f'eventIdx', []).append(event_id)
                            play_dict.setdefault(f'gameType', []).append(game_type)
                            play_dict.setdefault(f'gamePk', []).append(game)
                            play_dict.setdefault(f'player_name',[]).append(fn)
                            play_dict.setdefault(f'eventCode',[]).append(event_code)
                            
                    else:
                        pass 
                        
                        


        df = pd.DataFrame(play_dict)
        df['play_id'] = df['gamePk'].astype(str) + "_" + df['eventIdx'].astype(str)
        
        #Getting team_id of current game
        test_dict = {}
        for i in ['home', 'away']:
            team_id = j['liveData']['boxscore']['teams'][i]['team']['id']
            for k,v in j['liveData']['boxscore']['teams'][i]['players'].items():
                player = v['person']['id']
                player_name = v['person']['fullName']


                test_dict.setdefault(f'playerId', []).append(player)
                test_dict.setdefault(f'team_id', []).append(team_id)
            
        join_df = pd.DataFrame(test_dict)
        df = pd.merge(df, join_df, how = 'left', on = 'playerId')
        df_list.append(df)
    except KeyError:
        pass
```

```python
final_df = pd.concat(df_list)
```

```python
final_df = final_df[['play_id', 'gamePk', 'player_name','playerType', 'playerId', 'eventIdx', 'gameType', 'eventCode', 'team_id']]
```
